<br>
<h2 style="text-align: center; color: white">Putting Everything Together<b><b></h2>
    
- Early Stopping
    
<br>

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python

# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# scaling values
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# no. of input features 
input_shape = x_train.shape[1:]
input_shape 

# Neural Architecture
model = Sequential()
model.add(Dense(activation = 'relu', units = 14, input_dim = 13, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'relu', units = 8, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'relu', units = 4, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'sigmoid', units = 1, kernel_initializer = 'uniform'))

# summary
model.summary()

# compiling 
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

early_stopping = EarlyStopping(
    monitor='val_loss', # it monitors validation loss
    patience=15,         # these are Number of epochs to wait for improvement
    restore_best_weights=True)

# fitting
history = model.fit(x_train, y_train, batch_size = 16, epochs = 50, validation_split = 0.3, callbacks=[early_stopping])

# history
history.history

# y_pred
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)

# confusion_matrix
print(confusion_matrix(y_test, y_pred))

# classification_report
print(classification_report(y_test, y_pred))
```

<br>
<h2 style="text-align: center; color: white">Changing Number of Neurons to Get Higher Accuracy<b><b></h2>    
<br>

```python

# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# scaling values
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# no. of input features 
input_shape = x_train.shape[1:]
input_shape 

# Neural Architecture
model = Sequential()
model.add(Dense(activation = 'relu', units = 8, input_dim = 13, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'relu', units = 8, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'relu', units = 32, kernel_initializer = 'uniform'))
model.add(Dense(activation = 'sigmoid', units = 1, kernel_initializer = 'uniform'))

# summary
model.summary()

# compiling 
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

early_stopping = EarlyStopping(
    monitor='val_loss', # it monitors validation loss
    patience=30,         # these are Number of epochs to wait for improvement
    restore_best_weights=True)

# fitting
history = model.fit(x_train, y_train, batch_size = 8, epochs = 50, validation_split = 0.3, callbacks=[early_stopping])

# history
history.history

# y_pred
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)

# confusion_matrix
print(confusion_matrix(y_test, y_pred))

# classification_report
print(classification_report(y_test, y_pred))
```
